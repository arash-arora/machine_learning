```python
from sklearn.linear_model import LinearRegression #Imported LR
from sklearn import datasets as ds
import pandas as pd
from sklearn.model_selection import train_test_split as tts
```

```python
model = LinearRegression() #LR model
house = ds.load_boston()
print(house.DESCR)
```

```python
X_train, X_test, Y_train, Y_test = tts(x, y, test_size=0.2, random_state=28)
```

```python
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
model.fit(X_train,Y_train) #training
```

```python
yprediction = model.predict(X_test)
```

```python
yprediction
```

```python
Y_test
```

```python
from sklearn.metrics import mean_squared_error as mse
import numpy as np
mse1 = mse(Y_test, yprediction)
rmse = np.sqrt(mse1)
print("Root mean squared error:", rmse)
```
